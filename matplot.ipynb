```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_prop_23 = pd.read_excel('./DBs/df_srag_populacao_PRINCIPAIS_PRECISO_23.xlsx')
df_prop_24 = pd.read_excel('./DBs/df_srag_populacao_PRINCIPAIS_PRECISO_24.xlsx')
```

# Organização do Maior para o Menor

```python
df_prop_23_sorted = df_prop_23.sort_values(by='PROPORCAO', ascending=False).reset_index(drop=True)

df_prop_24_sorted = df_prop_24.sort_values(by='PROPORCAO', ascending=False).reset_index(drop=True)
df_prop_24_sorted.head(30)
```

```python
df_prop_23_sorted = df_prop_23_sorted.iloc[:19]
display(df_prop_23_sorted)

df_prop_24_sorted = df_prop_24_sorted.iloc[:20]
display(df_prop_24_sorted)

df_prop_23_sorted.info()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df_prop_23_sorted['CIDADE'], df_prop_23_sorted['PROPORCAO'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Casos por 100 mil habitantes')
plt.title('Proporção de casos por 100 mil habitantes por cidade')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(df_prop_24_sorted['CIDADE'], df_prop_24_sorted['PROPORCAO'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Casos por 100 mil habitantes')
plt.title('Proporção de casos por 100 mil habitantes por cidade')
plt.tight_layout()
plt.show()
```

```python
df_merged = pd.merge(
    df_prop_23_sorted[['CIDADE', 'PROPORCAO']],
    df_prop_24_sorted[['CIDADE', 'PROPORCAO']],
    on='CIDADE',
    suffixes=('_2023', '_2024')
)

# Criar o gráfico
fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(df_merged['CIDADE']))
width = 0.35

# Criar as barras
bars_2023 = ax.bar(x - width/2, df_merged['PROPORCAO_2023'], width, label='2023', color='skyblue')
bars_2024 = ax.bar(x + width/2, df_merged['PROPORCAO_2024'], width, label='2024', color='salmon')

# Adicionar os valores sobre as barras
for bar in bars_2023:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}', ha='center', va='bottom', fontsize=8)

for bar in bars_2024:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}', ha='center', va='bottom', fontsize=8)

# Ajustar rótulos e layout
ax.set_xlabel('Cidade')
ax.set_ylabel('Casos por 100 mil habitantes')
ax.set_title('Comparação da Proporção por Cidade - 2023 vs 2024')
ax.set_xticks(x)
ax.set_xticklabels(df_merged['CIDADE'], rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
