```python
# Casos proporcionalmente à população
df_alt = pd.read_csv('alt_df_srag_23.csv', encoding='latin1', sep=';')

populacao_dict = {
    'SAO PAULO': 11345200,
    'CUBATAO': 112476,
    'SANTOS': 433656,
    'SAO VICENTE': 329911,
    'PRAIA GRANDE': 319000,
    'MONGAGUA': 53100,
    'ITANHAEM': 112476,
    'BERTIOGA': 64188,
    'GUARUJA': 323000,
    'SAO BERNARDO DO CAMPO': 810729,
    'MOGI DAS CRUZES': 449955,
    'SUZANO': 302000,
    'POA': 103765,
    'ITAQUAQUECETUBA': 400000,
    'FERRAZ DE VASCONCELOS': 179198,
    'SANTO ANDRE': 718000,
    'SAO CAETANO DO SUL': 160000,
    'DIADEMA': 393237,
    'RIBEIRAO PIRES': 115559,
    'RIO GRANDE DA SERRA': 59000,
    'PAULINIA': 110537,
    'MAUA': 418261
}

populacao_df = pd.DataFrame(list(populacao_dict.items()), columns=['CIDADE', 'POPULACAO'])

srag_sp_df = df_alt[df_alt['SG_UF_NOT'].str.contains("SP", na=False)]

# Contar o número de casos por cidade (somar as ocorrências da cidade)
casos_por_cidade = srag_sp_df['ID_MUNICIP'].value_counts().reset_index()

casos_por_cidade.columns = ['CIDADE', 'QUANTIDADE DE CASOS']

# Merge com os dados de população
df_completo = pd.merge(casos_por_cidade, populacao_df, on='CIDADE', how='left')

# Calcular a proporção de SRAG por 100.000 habitantes
df_completo['PROPORCAO'] = (df_completo['QUANTIDADE DE CASOS'] / df_completo['POPULACAO']) * 100000
df_completo['PERCENTUAL_IMPACTADO'] = (df_completo['QUANTIDADE DE CASOS'] / df_completo['POPULACAO']) * 100

df_completo.columns = df_completo.columns.str.upper()

# DataFrame final
(df_completo)
df_completo.to_excel('./DBs/df_srag_populacao_PRINCIPAIS_PRECISO_23.xlsx', index=False)
```
