```python
import pandas as pd
import numpy as np

df_poluicao = pd.read_excel('./DBs/df_poluicao_regiao.xlsx', sheet_name=None)

df_poluicao
```

```python
# Lista das colunas que você quer tratar
colunas_para_tratar = [
    'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 
    'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro', 'Média do Período'
]

# Iterar por todas as planilhas (cada uma será um DataFrame)
for sheet_name, df in df_poluicao.items():
    print(f"Processando a planilha: {sheet_name}")
    
    # Substituir valores "-" por 0 e aplicar .strip() nas colunas específicas
    for col in colunas_para_tratar:
        if col in df.columns:
            df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) and x != '-' else (0 if isinstance(x, str) and x == '-' else x))
    
    # Converter colunas numéricas para int nas colunas específicas
    for col in colunas_para_tratar:
        if col in df.columns:
            # Verificar se a coluna é numérica (ignorando texto)
            df[col] = pd.to_numeric(df[col], errors='coerce')
            # Preencher NaN com 0 e converter para inteiro
            df[col] = df[col].fillna(0).astype(int)
    
    # Atualizar a planilha processada
    df_poluicao[sheet_name] = df  # Substituir a planilha original pelo tratado

# Para verificar os primeiros dados de cada planilha após o tratamento
for sheet_name, df in df_poluicao.items():
    print(f"\n{sheet_name} - Exemplo de dados após tratamento:")
    print(df.head())
```

```python
import pandas as pd

# Salvando o dicionário de DataFrames de volta em um arquivo Excel
with pd.ExcelWriter('./DBs/df_poluicao_regiao_filtrado.xlsx') as writer:
    for sheet_name, df in df_poluicao.items():
        # Salvar cada planilha no arquivo Excel
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
